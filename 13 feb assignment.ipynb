ValueError: It is raised when an operation or function receives an argument that has the right type but an inappropriate value.

```python
def calculate_price(unit_price, quantity):
    if unit_price <= 0:
        raise ValueError("Unit price must be greater than zero")
    if quantity <= 0:
        raise ValueError("Quantity must be greater than zero")
    total_price = unit_price * quantity
    return total_price

# Example usage
try:
    price = calculate_price(-10, 5)
except ValueError as e:
    print("Error:", e)
```
