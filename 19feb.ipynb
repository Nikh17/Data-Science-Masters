Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a open-source web development framework created in Python language.

Advantages of Flask:

It has an inbuilt development server.
It has vast third-party extensions.
It has a tiny API and can be quickly learned by a web developer.
It is WSGI compliant.
It supports Unicode.


Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook.

```python
from flask import Flask
app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello World'

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask? Why do we use app routes?

App routing refers to a mechanism of mapping HTTP requests to corresponding functions or views that will handle the request and return a response.

We use app routes in Flask to define the behavior of our web application in response to incoming requests from clients.

Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

```python
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def index():
    return "Welcome to ABC Corporation"


@app.route('/')
def index():
    return "Company Name: ABC Corporation,  Location: India,   Contact Detail: 999-999-9999"


if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'This is the home page.'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/user/<username>')
def user(username):
    return 'Welcome %s' % username

with app.test_request_context():
    print(url_for('home'))
    print(url_for('about'))
    print(url_for('user', username='John Doe'))
```
